```python
import pandas as pd
import numpy as np

pd.options.display.max_rows = 100
pd.options.display.max_columns = 100
```

```python
am = pd.read_pickle('./02_adjacency_matrix.pkl')
am[0:2]
```

```python
def affects_growth(common_plant_name):
    return am[common_plant_name][am[common_plant_name]!=0.0]

def growth_affected_by(common_plant_name):
    return am.loc[common_plant_name][am.loc[common_plant_name] != 0.0]
```

```python
def custom_merge(dict1, dict2):
    for k, v in dict1.items():
        if dict2.get(k):
            dict2[k] = dict2[k] + v
        else:
            dict2[k] = v
    
    return dict2
```

```python
custom_merge({"a": 2}, {"b": 2, "a": 1})
```

```python
def find_good_companions(plants, n=1, i=0):
    
    t_helped_by, t_helps = {}, {}
    t_hurt_by, t_hurts = {}, {}
    
    for p in plants:
        
        # helps
        helps = growth_affected_by(p)
        helped_by = affects_growth(p)    
        helps = helps[helps > 0].to_dict()
        helped_by = helped_by[helped_by > 0].to_dict()
        
        t_helped_by = custom_merge(helped_by, t_helped_by)
        t_helps = custom_merge(helps, t_helps)
        
        # hurts
        hurts = growth_affected_by(p)
        hurt_by = affects_growth(p)    
        hurts = hurts[hurts < 0].to_dict()
        hurt_by = hurt_by[hurt_by < 0].to_dict()
        
        t_hurt_by = custom_merge(hurt_by, t_hurt_by)
        t_hurts = custom_merge(hurts, t_hurts)
    
    return {'helped_by': t_helped_by, 'helps': t_helps, 
            'hurts': t_hurts, 'hurt_by': t_hurt_by}
    
    
    
```

```python
find_good_companions(['passion fruit', 'beets', 'brassicas'])
```
